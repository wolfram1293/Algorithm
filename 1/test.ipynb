```python
with open('test1.txt', mode='r') as f:
    y11 = list(map(float, f.read().split(',')))

with open('test1vm.txt', mode='r') as f:
    y12 = list(map(float, f.read().split(',')))
```

```python
import matplotlib.pyplot as plt
x = [10**i for i in range(1, 8)]
plt.figure()
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(x, y11, label='Native OS')
ax.plot(x, y12, label='Guest OS')
ax.set_xlabel("n")
ax.set_ylabel("time(s)")
ax.legend()
plt.savefig("test1.png", format="png", dpi=300)
plt.show()
```

```python
dy = [y12[i]/y11[i] for i in range(len(x))]
print(sum(dy)/len(x))
plt.figure()
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.plot(x, dy, label='ratio')
ax.set_xlabel("n")
ax.set_ylabel("ratio")
plt.savefig("test1d.png", format="png", dpi=300)
plt.show()
```
